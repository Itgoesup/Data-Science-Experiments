```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.info()
```

```python
tips.head()
```

```python
#Distribution plot for ONE Variable
#kde for Kernel Density Estimate
#bin is number of entries to show
sns.histplot(tips['total_bill'],kde=True,bins=40)
```

```python
#Combine two different distribution plots
#Compare two histograms, X column, Y Column, Dataframe
#kind changes the visuals
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
#Draws a regression line
g= sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
#Putting in hue will colour those with the colour of that category
#palette will add a colour palette
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#KDE Plots
sns.rugplot(tips['total_bill'])
```

```python
#Categorical Data
import seaborn as sns
%matplotlib inline
tips=sns.load_dataset('tips')
tips.head()
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
import numpy as np
```

```python
#Estimator function lets you enter numpy functions. Here we put np.std which
#shows us the standard deviation of both male and female
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
#Same as bar plot but only takes one category and counts
sns.countplot(x='sex',data=tips)
```

```python
#BoxPlot
#Power of boxplot is to see the comparison side by side with a category
#you choose.
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue='sex',dodge=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['sepal_length'].unique()
```

```python
g= sns.PairGrid(iris)
#map the plot type
g.map(plt.scatter)
```

```python
#Diagonal plot
g=sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
